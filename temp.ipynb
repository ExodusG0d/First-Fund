```python
import pandas as pd

data = pd.read_csv('NetValue.csv')
data.columns = ['Date', 'Strategy_Value', 'Benchmark_Value']

data['Strategy_Value'] = data['Strategy_Value']/data['Strategy_Value'].iloc[0]
data['Benchmark_Value'] = data['Benchmark_Value']/data['Benchmark_Value'].iloc[0]

data.to_csv('Normalized_NetValue.csv', index=False)
```

```python
import numpy as np

corr = np.corrcoef(data['Strategy_Value'], data['Benchmark_Value'])
print(corr)
```
